```python
import json
import plotly.express as px
import plotly

import pandas as pd 

from data_processing import preprocess_dataframe, melt_dataframe

df = pd.read_csv("Subnational_total_final_energy_consumption_statistics.csv")
dff = preprocess_dataframe(df)
dff = dff.rename(columns={"All_Fuels_Total": "Total (GWh)"})
with open("nuts_level_1.geojson") as injson:
    geojson = json.load(injson)
    
def update_choropleth(year_value):
    year_df = dff[dff["Year"] == year_value]
    fig = px.choropleth_mapbox(year_df, geojson=geojson, locations="Name", color="Total (GWh)", featureidkey="properties.nuts118nm",
                               range_color=(0, 316000), color_continuous_scale=plotly.colors.diverging.Temps, title=f"Energy consumption in the UK ({year_value})",
                              )
    fig.update_layout(mapbox_style="carto-positron",
                  mapbox_zoom=3.4, mapbox_center={"lat": 55.7, "lon": -5})
    fig.update_traces(textposition='inside', textinfo='label')

    return fig

# for year in range(2005, 2019):
#     fig = update_choropleth(year)    
#     fig.write_image(f"choropleth/{year}.jpeg")
fig = update_choropleth(2005)
fig.show()
```
