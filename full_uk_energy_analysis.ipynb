## Importing the libraries 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import plotly.express as px 
```

## Importing the dataset

```python
df = pd.read_csv("Subnational_total_final_energy_consumption_statistics.csv")
df
```

# Preprocessing the dataset

## Filtering rows with GWh units

```python
# Filter DataFrame based on values of the UNIT column
df = df[df["UNIT"] == "GWh"]
df
```

## Filtering only regions

```python
# Filter by values in all caps
regions_df = df[df["NAME"].str.isupper()]
regions_df
```

## Making the dataset easier to look at

It's not required but we're going to drop the LAUA column, reorder the columns, title the region and column names, and reset the index to make our dataset a little easier to look at

```python
# Reorder the columns 
descriptive_columns = ["YEAR", "NAME", "UNIT"]
other_columns = [col for col in regions_df.columns if col not in descriptive_columns]
all_columns = descriptive_columns + other_columns
regions_df = regions_df[all_columns]

# Drop the LAUA column
regions_df = regions_df.drop(columns=["LAUA"])

# Title the columns 
regions_df = regions_df.rename(columns={col: col.title() for col in descriptive_columns})

# Title the names
regions_df["Name"] = regions_df["Name"].str.title()

# Reset the index 
regions_df = regions_df.reset_index(drop=True)

regions_df
```
